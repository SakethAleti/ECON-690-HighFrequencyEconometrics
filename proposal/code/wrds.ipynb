```python
import pandas as pd
import numpy as np
import wrds
import sys
import datetime as dt



### Functions

def list_str(li):
    # Convert python list to a string representation 
    # of the list for SQL queries
    return ("'" + "', '".join(li) + "'")

def print_with_time(string):
    # Print function that prepends string with 
    # the current date and time
    print(f'[{dt.datetime.now()}]: ' + string)

    
    
### Main
    
# Params
query_year = '2020'
query_month = '03'
output_filename = 'sp500_001_100_5minprices'

# Log in
db = wrds.Connection(**{'wrds_username': 'sa400'})

## Symbol subset

# Symbols to query
sym_list = list(pd.read_csv('../data/keys/sp500_codes.txt', header = None).iloc[:,0])
sym_list_sql = list_str(sym_list[:2])

## Process Data
print_with_time(f'Starting data processing')

# For each table for the selected year
trade_tables_list = [x for x in db.list_tables(library = 'taqm_' + str(query_year)) 
                     if f'ctm_{query_year}{query_month}' in x and 'ix' not in x]

if not len(trade_tables_list):
    raise Exception(f'No observations for {query_year}-{query_month}')

# Five minute price dataframes
output_resample_df_list = []

for sql_table_name in trade_tables_list:

    ## Load Data
    print_with_time(f'    Collecting data from {sql_table_name}')
    
    # SQL Query
    raw_output_df= db.raw_sql(f'''
    SELECT date,cast(time_m as VARCHAR),ex,sym_root,price,
        (CAST(FLOOR((EXTRACT(hour from time_m)*60 + EXTRACT(minute from time_m))/5) AS INT)) AS mod_time_m
    FROM taqm_2020.{sql_table_name}
    WHERE (sym_suffix IS NULL)
        AND (sym_root IN ({sym_list_sql}))
    ''')
    
    # Check if empty
    if not len(raw_output_df):
        print_with_time(f'    No observations for  {sql_table_name}')
        continue

    ## Clean
    print_with_time(f'    Cleaning data for    {sql_table_name}')

    # Filter
    raw_output_df = raw_output_df.loc[(raw_output_df['mod_time_m'] >= 114) & (raw_output_df['mod_time_m'] <= 191),:]
    raw_output_df = raw_output_df.loc[(raw_output_df['ex'] != 'M') & (raw_output_df['ex'] != 'D'),:]

    # Fix dtypes to save space
    output_df = raw_output_df.astype(dtype={
            "sym_root": "string",
            "mod_time_m": 'int16'
        }).rename(columns = {'sym_root': 'ticker'})

    # Resample
    output_resample_df = (
        output_df.groupby(["date", "ticker", "mod_time_m"])[["time_m", "price"]].last().reset_index()
    )
    output_resample_df["datetime"] = pd.to_datetime(
        output_resample_df["date"].astype(str)
        + " "
        + output_resample_df["mod_time_m"].apply(
            lambda x: f"{int(np.floor((x+1)*5/60)):02d}:{(x+1)*5 % 60:02d}"
        )
    )

    output_resample_df_list.append(output_resample_df)
    
    # Save 
    query_date = str(output_resample_df['date'].iloc[0]).replace('-', '')
#     output_resample_df.to_csv(f'../../data/clean/daily/{output_filename}_{query_date}.csv', index = False)
    
# ## Aggregate resampled daily files and save
# print_with_time('Saving results')
# output_resample_agg_df = pd.concat(output_resample_df_list)
# output_resample_agg_df.to_csv(f'../../data/clean/monthly/{output_filename}_{query_year}{query_month}.csv', index = False)
```
